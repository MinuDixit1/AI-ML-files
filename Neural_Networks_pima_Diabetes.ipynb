#### Context :
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

Columns details are as below :

Pregnancies : Number of times pregnant    
Glucose : Plasma glucose concentration a 2 hours in an oral glucose tolerance test     
BloodPressure : Diastolic blood pressure (mm Hg)    
SkinThickness : Triceps skin fold thickness (mm)     
Insulin : 2-Hour serum insulin (mu U/ml)     
BMI : Body mass index (weight in kg/(height in m)^2)     
DiabetesPedigreeFunction : Diabetes pedigree function    
Age : Age (years)    
Outcome : Class variable (0 or 1) 268 of 768 are 1, the others are 0    
So, here the y-variable is Outcome

#### Create Model # 1 :   
Entire Dataset Used, Entire X is Normalized.    
Input Layer : 8 Neurons   
Hidden Layer 1 : 5 Neurons : Sigmoid  
Hidden Layer 2 : 5 Neurons : Sigmoid  
Output Layer : 1 Neuron : Sigmoid  
Optimizer : sgd

```python
# create model
# Any Hidden Layer having input dimension as paramter will be treated as the first hidden layer. The last hidden layer will be taken as Output Layer

model1 = Sequential(); 
model1.add(Dense(5, input_dim=8, activation='sigmoid')) # Input Layer : 8 Neurons, Dense Layer : 5 Neurons
model1.add(Dense(5, activation='sigmoid')) # Dense Layer : 5 Neurons
model1.add(Dense(1, activation='sigmoid')) #Output Layer : 1 Neuron, as it is Binary Class Problem
# Compile model
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model -Entire Data, No validation Data
model1.fit(X, Y, epochs=150, batch_size=10)
```

```python
#Manual Validation
# MLP with manual validation set
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy

# fix random seed for reproducibility
numpy.random.seed(7)


# load dataset
colnames = ['preg', 'glu', 'bp', 'sft', 'ins', 'bmi', 'dpf', 'age', 'outcome']
dataset = pd.read_csv("/content/drive/My Drive/Colab Notebooks/All Data Files/pima-indians-diabetes.data",names=colnames)

# split into input (X) and output (Y) variables
X=dataset[['preg', 'glu', 'bp', 'sft', 'ins', 'bmi', 'dpf', 'age']]
Y=dataset['outcome']

#splitting data to train(67%) and test(33%)
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.3,random_state=7)

# create model
model2 = Sequential()
model2.add(Dense(12, input_dim=8, activation='sigmoid'))
model2.add(Dense(8, activation='sigmoid'))
model2.add(Dense(1, activation='sigmoid'))

# Compile model
model2.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
# Fit the model

model2.fit(train_x, train_y, validation_data=(test_x,test_y), epochs=150, batch_size=10)
```

Here the Labels are converted to two classes using one-hot encoders, thereby eliminating the dependency on threshold values.

```python
#Manual Validation
# MLP with manual validation set
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy

# fix random seed for reproducibility
numpy.random.seed(7)

# load dataset
colnames = ['preg', 'glu', 'bp', 'sft', 'ins', 'bmi', 'dpf', 'age', 'outcome']
dataset = pd.read_csv("/content/drive/My Drive/Colab Notebooks/All Data Files/pima-indians-diabetes.data",names=colnames)

# split into input (X) and output (Y) variables
X=dataset[['preg', 'glu', 'bp', 'sft', 'ins', 'bmi', 'dpf', 'age']]
Y=dataset['outcome']

#splitting data to train(67%) and test(33%)
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.3,random_state=7)
train_y_hot = tf.keras.utils.to_categorical(train_y, num_classes=2)
test_y_hot = tf.keras.utils.to_categorical(test_y, num_classes=2)

# create model
model3 = Sequential()
model3.add(Dense(12, input_dim=8, activation='sigmoid'))
model3.add(Dense(8, activation='sigmoid'))
model3.add(Dense(2, activation='sigmoid'))

# Compile model
model3.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

# Fit the model
model3.fit(train_x, train_y_hot, validation_data=(test_x,test_y_hot), epochs=150, batch_size=10)
```

```python
 # evaluate the model
scores = model3.evaluate(test_x,test_y_hot)
scores[1]*100
```

```python
#Let's see the predicted values of the first record, here we will see the paired output (and not one) as there are two neurons in the output layer, one for each classes (one-hot encoded)
y_pred3= model3.predict(test_x[0:1])
print(y_pred3)
#Predict gives a continuous value, this has to be slashed to 0 and 1 using a threshold, for this simply use predit_classes
print(model3.predict_classes(test_x[0:1]))
```

```python
y_pred3= model3.predict(test_x)
out3 = []
for val in y_pred3:
  out3.append(np.argmax(val))
print(out3)
```

```python
from sklearn import metrics
cr3 = metrics.classification_report(test_y, out3)
print(cr3)
```

 Conclusion for Model 3 : 
 Training Loss is converging at : 0.61
